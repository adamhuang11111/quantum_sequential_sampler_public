```python
A1_A2_tab = {"rating": [], "type": [], "pred": []}
A1_A3_tab = {"rating": [], "type": [], "pred": []}
A2_A3_tab = {"rating": [], "type": [], "pred": []}

for j in range(data.shape[0]):
    for i in range(0, 78):
        if "A3" not in pred[i]:
            A1_A2_tab["rating"].append(data[j,i])
            A1_A2_tab["pred"].append(pred[i])
            A1_A2_tab["type"].append("observed")
            A1_A2_tab["rating"].append(BS_MS[j,i])
            A1_A2_tab["pred"].append(pred[i])
            A1_A2_tab["type"].append("BS_predicted")
            A1_A2_tab["rating"].append(PD_MS[j,i])
            A1_A2_tab["pred"].append(pred[i])
            A1_A2_tab["type"].append("PD_predicted")
            A1_A2_tab["rating"].append(RF_MS[j,i])
            A1_A2_tab["pred"].append(pred[i])
            A1_A2_tab["type"].append("RF_predicted")
        elif "A2" not in pred[i]:
            A1_A3_tab["rating"].append(data[j,i])
            A1_A3_tab["pred"].append(pred[i])
            A1_A3_tab["type"].append("observed")
            A1_A3_tab["rating"].append(BS_MS[j,i])
            A1_A3_tab["pred"].append(pred[i])
            A1_A3_tab["type"].append("BS_predicted")
            A1_A3_tab["rating"].append(PD_MS[j,i])
            A1_A3_tab["pred"].append(pred[i])
            A1_A3_tab["type"].append("PD_predicted")
            A1_A3_tab["rating"].append(RF_MS[j,i])
            A1_A3_tab["pred"].append(pred[i])
            A1_A3_tab["type"].append("RF_predicted")
        elif "A1" not in pred[i]:
            A2_A3_tab["rating"].append(data[j,i])
            A2_A3_tab["pred"].append(pred[i])
            A2_A3_tab["type"].append("observed")
            A2_A3_tab["rating"].append(BS_MS[j,i])
            A2_A3_tab["pred"].append(pred[i])
            A2_A3_tab["type"].append("BS_predicted")
            A2_A3_tab["rating"].append(PD_MS[j,i])
            A2_A3_tab["pred"].append(pred[i])
            A2_A3_tab["type"].append("PD_predicted")
            A2_A3_tab["rating"].append(RF_MS[j,i])
            A2_A3_tab["pred"].append(pred[i])
            A2_A3_tab["type"].append("RF_predicted")
```

```python
A1_A2_tab = pd.DataFrame(A1_A2_tab)
A1_A2_tab
```

```python
A1_A3_tab = pd.DataFrame(A1_A3_tab)
A2_A3_tab = pd.DataFrame(A2_A3_tab)
```

```python
def get_model_tab(model_name,pair_tab):
    return pair_tab[(pair_tab["type"] == model_name) || (pair_tab["ty[]"])]
```

```python
model_name = A1_A2_tab["type"].unique()
tab_list = [A1_A2_tab,A1_A3_tab,A2_A3_tab]
tab_name = ["A1_A2", "A1_A3", "A2_A3"]
```

```python
model_data = {"tab":{"A1_A2":{}, "A1_A3":{}, "A2_A3":{}}, "mean":{"A1_A2":{}, "A1_A3":{}, "A2_A3":{}}}
```

```python
for i in range(0,len(tab_name)):
    for n in model_name:
        model_tab = get_model_tab(n,tab_list[i])
        model_data["tab"][tab_name[i]][n] = model_tab
        means = []
        for p in model_tab['pred'].unique():
            pred_tab = model_tab[model_tab["pred"] == p]
            means.append(np.mean(pred_tab["rating"]))
        model_data["mean"][tab_name[i]][n] = means
```

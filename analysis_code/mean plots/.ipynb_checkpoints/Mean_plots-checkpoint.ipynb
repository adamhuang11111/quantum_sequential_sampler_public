```python
def get_model_tab(model_name,pair_tab):
    return pair_tab[pair_tab["type"] == model_name]
```

```python
model_name = A1_A2_tab["type"].unique()
tab_list = [A1_A2_tab,A1_A3_tab,A2_A3_tab]
tab_name = ["A1_A2", "A1_A3", "A2_A3"]
```

```python
model_data = {"tab":{"A1_A2":{}, "A1_A3":{}, "A2_A3":{}}, "mean":{"A1_A2":{}, "A1_A3":{}, "A2_A3":{}}}
```

```python
for i in range(0,len(tab_name)):
    for n in model_name:
        model_tab = get_model_tab(n,tab_list[i])
        model_data["tab"][tab_name[i]][n] = model_tab
        means = []
        for p in model_tab['pred'].unique():
            pred_tab = model_tab[model_tab["pred"] == p]
            means.append(np.mean(pred_tab["rating"]))
        model_data["mean"][tab_name[i]][n] = means
```

# Sum of P(A1), P(A2), P(A3) analysis

```python
PD_MS = sp.loadmat("IndDat_int1add_Ms.mat")['Ms_all'][0: 1162,:]
noint_BIC = np.add(0,sp.loadmat("IndDat_nointadd_nLLS.mat")['nLLS'][0: 1162,:])
int1_BIC = np.add(3.841,sp.loadmat("IndDat_int1add_nLLS.mat")['nLLS'][0: 1162,:])
```

```python
sumthree_tab = {"Gsquare_diff": list(np.subtract(int1_BIC.flatten(),noint_BIC.flatten())), "Diff":[], "Type":[]}
```

```python
for i in range(0,1162):
    ss = 0
    ssn = 0
    if sumthree_tab["Gsquare_diff"][i] < 0:
        sumthree_tab["Type"].append("significant")
    else:
        sumthree_tab["Type"].append("insignificant")
    for j in range(0,len(pred)):
        if (pred[j] == "P(A1)") or (pred[j] == "P(A2)") or (pred[j] == "P(A3)"):
            ss += PD_MS[i,j]
        elif (pred[j] == "P(¬A1)") or (pred[j] == "P(¬A2)") or (pred[j] == "P(¬A3)"):
            ssn += PD_MS[i,j]
    sumthree_tab["Diff"].append(np.abs(ss - ssn))
```

```python
sumthree_tab = pd.DataFrame(sumthree_tab)
sumthree_tab = sumthree_tab.sort_values("Diff", ascending=True)
sumthree_tab = sumthree_tab.reset_index(drop = False)
sumthree_tab
```

```python
nonsig_ord = sumthree_tab[sumthree_tab["Gsquare_diff"] >= 0]
sig_ord = sumthree_tab[sumthree_tab["Gsquare_diff"] < 0]
```

```python
plt.rcParams["figure.figsize"] = (12,6)
plt.bar(nonsig_ord.index,nonsig_ord["Diff"], color = "blue", label = "G^2 insignificant")
plt.bar(sig_ord.index,sig_ord["Diff"], color = "red", label = "G^2 significant")
plt.hlines(np.mean(sumthree_tab["Diff"]),0,1162,color = "green")
plt.grid()
plt.legend()
plt.annotate("mean = " + str(32.90) + ", std = " + str(37.88), 
             xy =(580, np.mean(sumthree_tab["Diff"])), xytext =(450, np.mean(sumthree_tab["Diff"]) + 50), arrowprops = dict(facecolor ='green', shrink = 0.05), color = "green")
plt.xlabel("subject")
plt.ylabel("$\sum_{n = 1}^{3}$ P(An) - P(¬An)")
plt.savefig("bar_diff.jpg", dpi = 500)
```

```python
sns.kdeplot(data=sumthree_tab, x="Diff",label = "All")
sns.kdeplot(data = sig_ord, x = "Diff", label = "Significant")
sns.kdeplot(data = nonsig_ord, x = "Diff", label = "Insignificant")
plt.legend()
plt.grid()
plt.savefig("difference.jpg", dpi = 500)
# plt.show()3
```

```python
def get_conj_disj_tab(index, conj_disj_index):
    conj_disj_tab = {"rating": [], "type": [], "pred": []}
    for j in range(0,24):
        i = conj_disj_index[j]
#         print(np.mean([data[index,i],data[index,i+24]]))
        conj_disj_tab["rating"].append(data[index,i])
        conj_disj_tab["pred"].append(pred[i])
        conj_disj_tab["type"].append("data")
        conj_disj_tab["rating"].append(PD_MS[index,i])
        conj_disj_tab["pred"].append(pred[i])
        conj_disj_tab["type"].append("quantum")
        conj_disj_tab["rating"].append(PD_noint_MS[index,i])
        conj_disj_tab["pred"].append(pred[i])
        conj_disj_tab["type"].append("classical")
    conj_disj_tab = pd.DataFrame(conj_disj_tab)
    data_tab = conj_disj_tab[conj_disj_tab["type"] == "data"]
    quantum_tab = conj_disj_tab[conj_disj_tab["type"] == "quantum"]
    classical_tab = conj_disj_tab[conj_disj_tab["type"] == "classical"]
    return data_tab,quantum_tab,classical_tab
```

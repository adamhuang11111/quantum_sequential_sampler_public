# Make Mean Plots

```python
mean_plot_tab = {"tab":{}, "mean":{}}
```

```python
for pair in mean_tab["pair"].unique():
    mean_plot_tab["tab"][pair] = {}
    mean_plot_tab["mean"][pair] = {}
    for m in mean_tab["type"].unique():
        model_tab = mean_tab[mean_tab["pair"] == pair]
        model_tab = model_tab[model_tab["type"] == m]
        mean_plot_tab["tab"][pair][m] = model_tab
        means = []
        for p in model_tab['pred'].unique():
            pred_tab = model_tab[model_tab["pred"] == p]
            #Compute Means
            means.append(np.mean(pred_tab["rating"]))
        mean_plot_tab["mean"][pair][m] = means
```

## Two Sigma Participant Mean

```python
two_sig_p = sp.loadmat("2sig_participant.mat")['participant_list'][0: 1162,:]
two_sig_p = two_sig_p.flatten()
two_sig_p = np.subtract(two_sig_p,1) #matlab is one indexing
```

```python
two_sig_tab = {"tab":{}, "mean":{}}
for pair in mean_tab["pair"].unique():
    two_sig_tab["tab"][pair] = {}
    two_sig_tab["mean"][pair] = {}
    for m in mean_tab["type"].unique():
        model_tab = mean_tab[mean_tab["pair"] == pair]
        model_tab = model_tab[model_tab["type"] == m]
        model_tab = model_tab[model_tab["subject"].isin(two_sig_p)]
        two_sig_tab["tab"][pair][m] = model_tab
        means = []
        for p in model_tab['pred'].unique():
            pred_tab = model_tab[model_tab["pred"] == p]
            #Compute Means
            means.append(np.mean(pred_tab["rating"]))
        two_sig_tab["mean"][pair][m] = means
```

# Main Manuscript Figure

```python
pred_list = mean_plot_tab["tab"]["A1_A2"]["BS"]["pred"].unique()
in_list = ["A", "B", "¬A", "¬B", "A ∩ B", "¬A ∩ ¬B", "A ∪ ¬B", '¬A ∪ ¬B', '¬B|A', '¬A|B']
exclusion_list = []
for p in pred_list:
    if p not in in_list:
        exclusion_list.append(p)
```

```python
def manuscript_plot(tab, exclusion_list):
    for p in exclusion_list:
        tab = tab[tab["pred"] != p]
    return tab
```

```python
def manuscript_mean(tab, exclusion_list):
    means = []
    for p in tab['pred'].unique():
        pred_tab = tab[tab["pred"] == p]
        #Compute Means
        means.append(np.mean(pred_tab["rating"]))
    return means
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,4), sharey=True)
plt.rcParams["figure.figsize"] = (20,12)
#two sigma part
two_sig_tab_m_data = manuscript_plot(two_sig_tab["tab"]["A1_A2"]["data"],exclusion_list)
two_sig_tab_m_CSS = manuscript_plot(two_sig_tab["tab"]["A1_A2"]["CSS"],exclusion_list)
two_sig_tab_m_QSS = manuscript_plot(two_sig_tab["tab"]["A1_A2"]["QSS"],exclusion_list)
sns.barplot(x="pred", y="rating", data=two_sig_tab_m_data,
            alpha=0.5, palette = "ch:s=.25,rot=-.25", ci = 95, ax = axs[0])
trans = axs[0].transData
axs[0].scatter(two_sig_tab_m_CSS["pred"].unique(), 
               manuscript_mean(two_sig_tab_m_CSS,exclusion_list), 
            marker = "v", color = "orange", s = plt.rcParams['lines.markersize'] * 10, label = "Classical Sequential Sampler",
              transform=trans+offset(-6))
axs[0].scatter(two_sig_tab_m_QSS["pred"].unique(), 
               manuscript_mean(two_sig_tab_m_QSS,exclusion_list),
            marker = "^", color = "blue", s = plt.rcParams['lines.markersize'] * 10, label = "Quantum Sequential Sampler",
              transform=trans+offset(6))
axs[0].set_ylim(-5,105)
axs[0].set_xticklabels(axs[0].get_xticklabels(),rotation = 90)
axs[0].set_title("p < 0.05 (281 counts)")
#four sigma part
four_sig_tab_m_data = manuscript_plot(four_sig_tab["tab"]["A1_A2"]["data"],exclusion_list)
four_sig_tab_m_CSS = manuscript_plot(four_sig_tab["tab"]["A1_A2"]["CSS"],exclusion_list)
four_sig_tab_m_QSS = manuscript_plot(four_sig_tab["tab"]["A1_A2"]["QSS"],exclusion_list)
sns.barplot(x="pred", y="rating", data=four_sig_tab_m_data,
            alpha=0.5, palette = "ch:s=.25,rot=-.25", ci = 95, ax = axs[1])
trans = axs[1].transData
axs[1].scatter(four_sig_tab_m_CSS["pred"].unique(), 
               manuscript_mean(four_sig_tab_m_CSS,exclusion_list), 
            marker = "v", color = "orange", s = plt.rcParams['lines.markersize'] * 10, label = "Classical Sequential Sampler",
              transform=trans+offset(-6))
axs[1].scatter(four_sig_tab_m_QSS["pred"].unique(), 
               manuscript_mean(four_sig_tab_m_QSS,exclusion_list),
            marker = "^", color = "blue", s = plt.rcParams['lines.markersize'] * 10, label = "Quantum Sequential Sampler",
              transform=trans+offset(6))
axs[1].set_xticklabels(axs[1].get_xticklabels(),rotation = 90)
axs[1].set_ylim(-5,105)
axs[1].set_title("p < 0.00008 (47 counts)")
plt.subplots_adjust(bottom=0.22)
axs[0].spines["top"].set_visible(False)
axs[0].spines["right"].set_visible(False)
axs[1].spines["top"].set_visible(False)
axs[1].spines["right"].set_visible(False)
axs[1].spines["left"].set_visible(False)
axs[1].set_yticks([])
axs[1].set_ylabel("")
plt.tight_layout()
handles0, labels0 = axs[0].get_legend_handles_labels()
handles1, labels1 = axs[1].get_legend_handles_labels()
handles = [handles0[0], handles0[1]]
labels = [labels0[0], labels0[1]]
fig.legend(handles, labels, loc='upper left', fontsize = 11)
axs[0].set_xlabel("")
axs[1].set_xlabel("")
axs[0].set_ylabel("Probability Judgment",fontsize = 11)
fig.subplots_adjust(top=1)
axs[0].set_xticklabels(labels = axs[0].get_xticklabels(),fontsize = 12)
axs[1].set_xticklabels(labels = axs[1].get_xticklabels(),fontsize = 12)
fig.suptitle("A=Ohio, B=Michigan", size = 15, y =1.2)
plt.savefig("css_qss_manuscript_mean.jpg",dpi = 500)
```
